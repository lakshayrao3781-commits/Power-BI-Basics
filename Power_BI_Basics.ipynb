Q-1 : What is Power BI and why is it used in businesses?

Answer : Power BI is a Business Intelligence (BI) and data visualization tool developed by Microsoft. It helps businesses connect, analyze, and visualize data from different sources.

Why businesses use Power BI:
Data Visualization- Creates charts, graphs, dashboards, and reports.
Better Decision Making- Helps managers make data-driven decisions.
Real-Time Insights- Shows live business performance.
Data Integration - Connects with Excel, SQL Server, APIs, cloud services, etc.
Easy Sharing - Reports can be shared across teams and organizations.

Q-2 : Name and explain the three main components of Power BI

Answer : 1. Power BI Desktop -
A desktop application used to create reports and dashboards.
Used for data cleaning, modeling, and visualization.
2. Power BI Service -
Cloud-based online service.
Used to publish, share, and collaborate on reports.
3. Power BI Mobile -
Mobile app for Android/iOS.
Allows users to access reports and dashboards on mobile devices.

Q-3  : Explain the Power BI workflow ?

Answer: 1- Connect Data
Import data from Excel, SQL, CSV, APIs, etc.
2- Transform Data
Clean and modify data using Power Query.
3-Model Data
Create relationships between tables.
4-Create Visualizations
Build charts, dashboards, KPIs, and reports.
5-Publish Reports
Upload reports to Power BI Service.
6-Share & Collaborate
Share dashboards with team members.
7-Monitor & Refresh
Schedule automatic data refresh.

Q6  : Define Data View, Report View, and Model View. Explain the purpose of each view.

Answer : 1. Data View
Purpose:
Used to view table data row by row.
Helps inspect and validate data.
Use Cases:
Check missing values
Create calculated columns.

2. Report View
Purpose:
Used to create visualizations and dashboards.
Use Cases:
Add charts, slicers, tables, KPIs
Design reports.

3. Model View
Purpose:
Used to manage relationships between tables.
Use Cases:
Create table relationships
Manage star schema
Optimize data model

Question 8 : Split Owner Name to create two new columns as First Name and Last Name.

Answer : Steps in Power Query:
1.Open Transform Data.

2.Select the Owner_Name column.

3.Go to:
Transform → Split Column

4.Choose:
By Delimiter

5.Select:
Space " "

6.Split at:
Left-most delimiter

7.Click OK

8.Rename columns:
Owner_Name.1 → First_Name
Owner_Name.2 → Last_Name
